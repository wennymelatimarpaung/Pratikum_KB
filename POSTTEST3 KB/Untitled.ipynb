```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("games.csv")
```

```python
df.head()
```

### Kesimpulan : dari bar plot ini menampilkan data dengan title Opening ply victory status. Dimana x label = Victory Status dan y label =Opening Ply. maka output yang dihasilkan adalah Victory Status seperti  outiftime dengan rata-rata(opening ply 30), resign dengan rata-rata sama dengan outiftimenya (opening ply 30), mates dengan rata-rata (opening ply 20), draw dengan rata-rata sama seperti mate (opening ply 20).

```python
x = list(df["victory_status"])
y = list(df["opening_ply"])

plt.bar(x,y,color='red')
plt.xlabel("Victory Status")
plt.ylabel("Opening Ply")
plt.title("Opening ply victory status")
```

### Kesimpulan: Dari pie chart ini dapat di simpulkan bahwa pie chart tersebut menampilkan persentase perbandingan pemenang dalam bentuk grafik pie. Dimana victory status yang paling banyak yaitu white dengan jumlah persentasekan 49.9%. kemudian black memegang pemenang kedua diamana 45.4%. lalu draw paling sedikit 4.7%. Di pie chart ini tidak bisa menggunakan x labell dan y label.

```python
data_pie = df["winner"].value_counts().rename_axis("winner").reset_index(name="id_counts")
```

```python
plt.figure(figsize=(10,10), dpi = 300)
plt.pie(data_pie["id_counts"],labels=data_pie['winner'],startangle=90,autopct='%.1f%%')
plt.title('Persentase Perbandingan Pemenang')
plt.show()
```
